```python
def get_time_bars(df: pd.DataFrame, resolution: str = "S", num_units: int = 1):
    result = [] 
    tick_num = 0
    prev_price = None 
    tick_diff = 0
    prev_tick_rule = None 
    open_price = None

    cum_statistics = {
        "cum_ticks": 0,
        "cum_dollar_value": 0,
        "cum_volume": 0,
        "cum_buy_volume": 0 
    }
    high_price, low_price = - np.inf, np.inf 

    threshold = int(1e9) * num_units * {
        'D': 60 * 60 * 24,
        'H': 60 * 60,
        'MIN': 60,
        'S': 1 
    }[resolution]
    
    timestamp = None 

    cols = [
        "date_time",
        "tick_num",
        "instrument",
        "open",
        "high",
        "low",
        "close",
        "volume",
        "cum_buy_volume",
        "cum_ticks",
        "cum_dollar_value",
    ]

    for index, row in df.iterrows():
        # print(row)
        price = np.float64(row['Price'])
        tick_num += 1
        volume = row['Volume']
        dollar_value = price * volume

        tick_diff = price - prev_price if prev_price is not None else 0
        signed_tick = np.sign(tick_diff) if tick_diff != 0 else prev_tick_rule
        prev_tick_rule = signed_tick
        prev_price = price
        
        date_time = row['datetime']
                
        # round tick boundary up 
        timestamp_threshold = (( date_time.as_unit('ns').value + (threshold - 1) ) // threshold) * threshold 

        if timestamp is None:
            timestamp = timestamp_threshold
            
        if timestamp < timestamp_threshold:
            # generate a new bar 
            result.append(
                [
                    timestamp,
                    tick_num,
                    row['Symbol'], 
                    open_price,
                    max(high_price, open_price),
                    min(low_price, open_price),
                    price, # close price 
                    cum_statistics['cum_volume'],
                    cum_statistics['cum_buy_volume'],
                    cum_statistics['cum_ticks'],
                    cum_statistics['cum_dollar_value']
                ]
            )
            open_price = None
            high_price, low_price = -np.inf, np.inf
            cum_statistics = {
                "cum_ticks": 0,
                "cum_dollar_value": 0,
                "cum_volume": 0,
                "cum_buy_volume": 0 
            }
            timestamp = timestamp_threshold
            
        if open_price is None:
            open_price = price 
            
        high_price, low_price = max(high_price, price), min(low_price, price)
        cum_statistics['cum_ticks'] += 1
        cum_statistics['cum_dollar_value'] += dollar_value
        cum_statistics['cum_volume'] += volume
        
        if signed_tick == 1:
            cum_statistics['cum_buy_volume'] += volume
    
    return pd.DataFrame(result, columns=cols)
```

```python
def get_rolled_series(df, match_end=True):
    # series.set_index('date_time', inplace=True)
    roll_dates = df['Symbol'].drop_duplicates(keep='first').index
    gaps = df['Price'] * 0 
    iloc = list(df.index)
    iloc = [ iloc.index(i) - 1 for i in roll_dates ] 
    print(iloc)
    
    gaps.loc[roll_dates[1:]] = df['Price'].loc[roll_dates[1:]] - df['Price'].iloc[iloc[1:]].values 
    gaps = gaps.cumsum()
    if match_end: gaps -= gaps.iloc[-1] # roll backward 
    return gaps

trade_df['gaps'] = get_rolled_series(trade_df)
trade_df.head()
# trades_df.iloc[[208, 209]]
```
